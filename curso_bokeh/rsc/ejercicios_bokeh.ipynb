## Ejercicio 1

- Usando Bokeh intenta generar una gráfica que:
    - Visualice 5 marcas situadas en [x, y] diferentes (numpy.random)
    - Utilicen un radio diferente de entre 10 y 30
    - Utilice asteriscos como marcas
    - El color sea “firebrick”
    - El ancho de la linea sea 3

```python
from numpy.random import random

# create a new plot using figure
p = figure(plot_width=500, plot_height=500)

x = random(5) 
y = random(5) 
size = range(10, 31, 5)

# add a square renderer with a size, color, alpha, and sizes
p.asterisk(x, y, size=size, color="firebrick", line_width=3)

show(p) # show the results
```

## Ejercicio 4

- Repite el mismo plot, esta vez utilizando bokeh.plotting en vez de Charts

```python
from bokeh.sampledata.iris import flowers as df

# create a new plot using figure
p = figure(plot_width=600, plot_height=600, title='Datos de Iris')

size=10
alpha=0.5

p.circle(df[df.species == "setosa"].petal_length, 
          df[df.species == "setosa"].petal_width, 
          size=size, color="red", alpha=alpha, legend="setosa")

p.square(df[df.species == "versicolor"].petal_length, 
          df[df.species == "versicolor"].petal_width, 
          size=size, color="green", alpha=alpha, legend="versicolor")

p.triangle(df[df.species == "virginica"].petal_length, 
          df[df.species == "virginica"].petal_width, 
          size=size, color="blue", alpha=alpha, legend="virginica")

p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"
p.legend.location = "top_left"

show(p) # show the results
```

## Ejercicio 5

- Usando gridplot crea un Scatter Plot Matrix del dataset Iris:

```python
from bokeh.io import gridplot

def scatter_chart(df, x, y):
    return Scatter(df, x=x, y=y, color='species', marker='species' width=200, height=200)

def scatter_figure(df, x, y):
    p = figure(width=200, height=200)

    size=5
    alpha=0.5

    p.circle(df[df.species == "setosa"][x], 
              df[df.species == "setosa"][y], 
              size=size, color="red", alpha=alpha)

    p.square(df[df.species == "versicolor"][x], 
              df[df.species == "versicolor"][y], 
              size=size, color="green", alpha=alpha)

    p.triangle(df[df.species == "virginica"][x], 
              df[df.species == "virginica"][y], 
              size=size, color="blue", alpha=alpha)
    
    return p


plots = []
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
for x in columns:
    for y in columns:
        plots.append(scatter_chart(df, x, y))

p = gridplot([plots[0:4], plots[4:8], plots[8:12], plots[12:16]])

show(p)
```
